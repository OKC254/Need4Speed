```python
%cd {HOME}

import cv2
from IPython.display import HTML
from base64 import b64encode

video_path = '/content/runs/detect/predict3/trafficvid.avi'

# Read the video using OpenCV
cap = cv2.VideoCapture(video_path)

# Get the video dimensions
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

frame_num = int(cap.get(cv2.CAP_PROP_FPS))

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
writer = cv2.VideoWriter('output', fourcc, frame_num, (width, height))

# Loop through the frames and display them in the notebook
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Write the frame to the output file
    writer.write(frame)

    # Get the current frame as JPEG data
    _, jpeg = cv2.imencode('.jpg', frame)

    # Encode the JPEG data as base64 and create a data URI
    data_uri = 'data:image/jpeg;base64,' + b64encode(jpeg).decode('utf-8')

    # Display the video frame in the notebook
    display(HTML('<img src="%s" class="blog-image">' % data_uri))

# Release the video capture and writer objects
cap.release()
writer.release()
```
